# Import analysis and visualization libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

# Reading the dataset

```python
ball_by_ball=pd.read_csv('IPL Ball-by-Ball 2008-2020.csv')
```

```python
ball_by_ball.head()
```

```python
ball_by_ball.tail()
```

```python
ball_by_ball.info()
```

```python
ball_by_ball.describe()
```

```python
ball_by_ball.shape
```

```python
num_null = ball_by_ball.isnull().sum().sum()
```

```python
num_null
```

```python
plt.figure(figsize=(18, 18))

sns.displot(
data = ball_by_ball.isna().melt(value_name="Null Values"),
y = "variable",
hue = "Null Values",
multiple="fill",
aspect=3,
palette='viridis' 
)
plt.title('Non-Missing Values in Ball by Ball data', size = 18, color = 'black')
plt.xlabel(" ")
plt.ylabel(" ")
plt.xticks(size = 12, weight = 'bold', color = 'black')
plt.yticks(size = 12, weight = 'bold', color = 'black');
```

```python
correlation=ball_by_ball.corr()
```

```python
plt.figure(figsize=(18, 18))
sns.heatmap(correlation, xticklabels=correlation.columns, yticklabels=correlation.columns, annot=True)
```

# Most Run scored analysis

```python
most_runs=ball_by_ball.groupby(['id'])['total_runs'].sum().reset_index()
```

```python
most_runs.head()
```

```python
# arrange them by ascending order

asc_most_runs = most_runs.sort_values(by="total_runs", ascending=True)
asc_most_runs
```

```python
# Descending Order

desc_most_runs = most_runs.sort_values(by="total_runs", ascending=False)
desc_most_runs
```

```python
sns.pairplot(asc_most_runs)
```

# Matches dataset

```python
matches=pd.read_csv('IPL Matches 2008-2020.csv')
```

```python
matches.head()
```

```python
matches.tail()
```

```python
matches.info()
```

```python
matches.describe()
```

```python
# Add yearwise season column
matches['season'] = pd.DatetimeIndex(matches['date']).year
```

```python
matches.head()
```

```python
matches.isnull().sum()
```

```python
num_matches_null = matches.isnull().sum().sum()
print("Matches have {} null values.".format(num_matches_null))
```

```python
# as we see method column has most null values
final = matches.dropna(subset=['method'])
```

```python
final.head()
```

```python
final.describe()
```

```python
final.info()
```

```python
m_corr = final.corr()
plt.figure(figsize=(18, 18))
sns.heatmap(m_corr, xticklabels=m_corr.columns, yticklabels=m_corr.columns, annot=True)
```

```python
plt.figure(figsize=(17, 17))
sns.displot(
data=matches.isna().melt(value_name="Null Values"),
y="variable",
hue="Null Values",
palette="viridis",
multiple="fill",
aspect=3
)
plt.title('Non-Missing Values in Matches data', weight = 'bold', size = 20, color = 'black')
plt.xlabel(" ")
plt.ylabel(" ")
plt.xticks(size = 12, weight = 'bold', color = 'maroon')
plt.yticks(size = 12, weight = 'bold', color = 'maroon');
```
